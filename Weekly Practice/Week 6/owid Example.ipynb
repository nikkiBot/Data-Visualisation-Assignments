### owid Data acquisition and working with multi-index dataframe.

```python
from owid import catalog
```

```python
table = catalog.find('covid')
```

```python
df = table.iloc[0].load() # table is a list of tables, so we need to select the first one and load it
df
```

```python
df.loc[("IND",slice(None))] # slice(None) is equivalent to : in pandas, meaning this is all columns with the IND row index
```

```python
df1 = df.loc[("IND",slice(None))]
df1.head()
```

```python
df2 = df1.dropna(subset=["total_deaths"])
df2.head()
```

```python
l0 = df.index.get_level_values(0)
l0
```

```python
l1 = df.index.get_level_values(1)
l1
```
