## Data Acquisition 

```python
api =  'https://data.telangana.gov.in/api/1/metastore/schemas/dataset/items/62a1cc18-7613-460b-a0b1-4b71c78fa1ce'
r = requests.get(api) 
print(r.status_code) 
if r.status_code == 200:
    print("Success! \n")
data = r.json()
if data :
    print("Sucessfully fetched data from API and converted to JSON \n")
# print(data) -> Giving an error ConnectionError: ('Connection aborted.', ConnectionResetError(10054, 'An existing connection was forcibly closed by the remote host', None, 10054, None))
```

```python
# Construct df_max and df_min from the url in metadata and cache them in csv files


df_max = pd.read_csv(data['distribution'][0]['downloadURL'])
# df_max.head()
df_min = pd.read_csv(data['distribution'][1]['downloadURL'])
# df_min.head()
```

```python
df_max.head()
```

```python
# Rename Mandal to Dmcode and Mandal.1 to Mandal
df_max.rename(columns = {'Mandal':'Dmcode', 'Mandal.1':'Mandal'}, inplace = True)
df_max.head()
```

```python
df_min.head()
```

```python
months = df_max.columns[3:]
months
```

```python

# merged_df = df_max.merge(df_min, on = ['Dmcode','District', 'Mandal'], suffixes=('_max', '_min'))

# merged_df.head()

combined_df = pd.DataFrame()
for month in months:
    combined_df[month] = list(zip(df_max[month], df_min[month])) # zip returns a single tuple for each row

combined_df["Dmcode"] = df_max["Dmcode"]
combined_df["Mandal"] = df_max["Mandal"]
combined_df["District"] = df_max["District"]

# reordering columns
cols = combined_df.columns.tolist()
cols = cols[-3:] + cols[:-3] # last 3 columns + rest of the columns
combined_df = combined_df[cols]
# combined_df.head()
combined_df

```

```python
new_df = pd.DataFrame()

# Add multi level index "min" and "max" to each month:

for column in combined_df.columns[3:]:
    new_df[(column, 'max')] = combined_df[column].apply(lambda x: x[0])
    new_df[(column, 'min')] = combined_df[column].apply(lambda x: x[1])

new_df["Dmcode"] = combined_df["Dmcode"]
new_df["Mandal"] = combined_df["Mandal"]
new_df["District"] = combined_df["District"]

# reordering columns
cols = new_df.columns.tolist()
cols = cols[-3:] + cols[:-3] # last 3 columns + rest of the columns
new_df = new_df[cols]

warnings.filterwarnings('ignore')
new_df
```
