```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import matplotlib as plt
```

```python
# Reading Gender and Height data from excel file
df = pd.read_excel('200 participant collected data.xlsx', usecols = ['Gender', 'Height (cm) (average of 3 measurments)'])
```

```python
# Rename the height column to height for convenience
df.rename(columns={'Height (cm) (average of 3 measurments)':'Height'}, inplace=True)
df.head()
```

# DATA CLEANING

```python
# The gender column has some non-uniformity on the name. For example : "Male" may sometimes appear as "male" in some rows. Use pandas functions to fix this issue.
# string.capitalize() can be used to capitalize the first letter of a string.
df["Gender"] = df["Gender"].str.capitalize()
df 
```

# DATA VISUALISATION

```python
# Use seaborn to create histogram, box and violin plots. Note that distribution of male and female heights should be shown separately in the same plot (as they may overlap)
```

### Histogram

```python
x = sns.histplot(data=df,x="Height", hue="Gender", alpha=0.4)
x.set(xlabel = "Height (cm) (average of 3 measurments)")
```

### Box Plot

```python
box = sns.boxplot(data=df, x="Height", y="Gender", hue="Gender")
box.set(xlabel="Height (cm) (average of 3 measurments)")
```

### Violin Plot

```python
y = sns.violinplot(data=df, x="Height", y="Gender", hue="Gender")
y.set(xlabel="Height (cm) (average of 3 measurments)")
y.legend(loc='lower right')
```
